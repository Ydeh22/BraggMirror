```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pi= np.pi
```

```python
n1 = 1.0 #refracive index for air
n2= 1.5  #refracive index for glass

def Rs(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)

    return (np.abs(
        (n1 * np.cos(thI) - n2 * np.cos(thT)) /
        (n1 * np.cos(thI) + n2 * np.cos(thT))
    ))**2

def Rp(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)

    return (np.abs(
        (n1 * np.cos(thT) - n2 * np.cos(thI)) /
        (n1 * np.cos(thT) + n2 * np.cos(thI))
    ))**2
```

```python
for x in np.linspace(0,90,10):
    print(Rs(x*pi/180))
#some values for Rs
```

```python
for x in np.linspace(0,90,10):
    print(Rp(x*pi/180))
#some values for Rp
```

```python
print(Rp(31*pi/180))
```

```python
n1=1
n2=2

x = np.linspace(0,90,1000)

plt.plot(x, Rs(x*pi/180), linewidth=2)
plt.plot(x, Rp(x*pi/180), linewidth=2)

plt.xticks([0,15,30,45,60,75,90])

plt.xlim(0,90)
plt.ylim(0,1)

plt.xlabel('Angle of Incidence (degrees)')
plt.ylabel('Reflectivity')
#plt.title()

plt.legend(['Rs','Rp'])
plt.grid()


plt.savefig('Single Interface12')
plt.show()
```

```python
#now going from glass to air
n1=1.5
n2=1

x = np.linspace(0,90,1000)

plt.plot(x, Rs(x*pi/180))
plt.plot(x, Rp(x*pi/180))

plt.xticks([0,15,30,45,60,75,90])

plt.xlim(0,90)
plt.ylim(0,1)

plt.xlabel('Angle of Incidence (degrees)')
plt.ylabel('Proportion of the Incident Light Reflected (%)')
plt.title('Light Travelling from Glass into Air ')

plt.legend(['Rs','Rp'])
plt.grid()

plt.show()
```

```python
#the refelction amplitude coefficients for air to glass
n1=1
n2=1.5

def rs(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return (n1*np.cos(thI) - n2*np.cos(thT)) /  (n1*np.cos(thI) + n2*np.cos(thT))

def rp(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)

    return (n2 * np.cos(thI) - n1 * np.cos(thT)) / (n2 * np.cos(thI) + n1 * np.cos(thT))
    
```

```python
#some values of rs and rp
for i in np.linspace(0,90,10):
    print(rs(i*pi/180))
```
