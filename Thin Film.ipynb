```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pi= np.pi
k0 = 12566370 #for light of wavelength 500nm
```

```python
def rp01(thI):
    thT = np.arcsin( (n0*np.sin(thI)) /n1)
    
    return (n1 * np.cos(thI) - n0 * np.cos(thT)) / (n1 * np.cos(thI) + n0 * np.cos(thT))

def rp12(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return (n2 * np.cos(thI) - n1 * np.cos(thT)) / (n2 * np.cos(thI) + n1 * np.cos(thT))

def rs01(thI):
    thT = np.arcsin( (n0*np.sin(thI)) /n1)
    
    return (n0*np.cos(thI) - n1*np.cos(thT)) /  (n0*np.cos(thI) + n1*np.cos(thT))

def rs12(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return (n1*np.cos(thI) - n2*np.cos(thT)) /  (n1*np.cos(thI) + n2*np.cos(thT))


def rp_slab1(thI):
    return (rp01(thI) + rp12(thI)) / (1 + rp01(thI)*rp12(thI))

def rp_slab2(thI):
    return (rp01(thI) - rp12(thI)) / (1 - rp01(thI)*rp12(thI))


def rs_slab1(thI):
    return (rs01(thI) + rs12(thI)) / (1 + rs01(thI)*rs12(thI))

def rs_slab2(thI):
    return (rs01(thI) - rs12(thI)) / (1 - rs01(thI)*rs12(thI))
```

```python
def ts_slab(thI):
    return 1 + rs_slab(thI)
```

```python
def kx(thI):
    
    return k0*np.sin(thI)
```

```python
n0=1 #air
n1=2.0
n2=1.5  #glass

x = np.linspace(0,90,1000)

plt.plot(kx(x*pi/180), rp_slab1(x*pi/180))
plt.plot(kx(x*pi/180), rs_slab1(x*pi/180))
plt.plot(kx(x*pi/180), rp_slab2(x*pi/180))
plt.plot(kx(x*pi/180), rs_slab2(x*pi/180))


#plt.plot(kx(x*pi/180), tp_slab(x*pi/180))
#plt.plot(kx(x*pi/180), ts_slab(x*pi/180))

plt.xlim(0,)
plt.ylim(-1.2,1.4)

plt.title('Light Travelling from Air into Glass with a thin film ')
plt.ylabel('Amplitude Reflection & Transmission Coefficients')
plt.xlabel('X component of the WaveVector K')

plt.legend(['rp (+)','rs (+)','rp (-)','rs (-)'])
plt.grid()

plt.show()
```
