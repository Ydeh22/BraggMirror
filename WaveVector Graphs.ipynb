```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pi= np.pi
```

```python
k0 = 12566370 #for light of wavelength 500nm
n1=1
n2=1.5

def rp(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return (n2 * np.cos(thI) - n1 * np.cos(thT)) / (n2 * np.cos(thI) + n1 * np.cos(thT))

def rs(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return (n1*np.cos(thI) - n2*np.cos(thT)) /  (n1*np.cos(thI) + n2*np.cos(thT))

def ts(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return 2*n1*np.cos(thI) / (n1*np.cos(thI) + n2*np.cos(thT))

def tp(thI):
    thT = np.arcsin( (n1*np.sin(thI)) /n2)
    
    return 2*n1*np.cos(thI) / (n2*np.cos(thI) + n1*np.cos(thT))

def kx(thI):
    
    return k0*np.sin(thI)
```

```python
x = np.linspace(0,90,1000)

plt.plot(kx(x*pi/180), rp(x*pi/180))
plt.plot(kx(x*pi/180), rs(x*pi/180))

plt.plot(kx(x*pi/180), tp(x*pi/180))
plt.plot(kx(x*pi/180), ts(x*pi/180))

plt.xlim(0,)
plt.ylim(-1.2,1)

plt.title('Light Travelling from Air into Glass ')
plt.ylabel('Amplitude Reflection & Transmission Coefficients')
plt.xlabel('X component of the WaveVector K')

plt.legend(['rp','rs','tp','ts'])
plt.grid()

plt.show()
```
